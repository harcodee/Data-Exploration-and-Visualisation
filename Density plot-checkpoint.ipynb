```python
#Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#create a data frame
import pandas as pd

data = {
    'content': [
        "Hello, how are you?",
        "Please find the attachment.",
        "Thank you!",
        "Looking forward to your response.",
        "This is a test email with several words in it.",
        "Urgent: Meeting rescheduled to 3 PM.",
        "Can we connect sometime this week?",
        "Happy to help!",
        "Let me know your thoughts.",
        "Good morning!",
        "Your order has been shipped.",
        "Payment successful. Thank you!",
        "Letâ€™s catch up soon.",
        "The event is postponed.",
        "Important update: Read now.",
        "Lunch at 1 PM?",
        "Call me when you're free.",
        "Here's the file you asked for.",
        "New login to your account detected.",
        "Welcome to our service!"
    ]
}

df = pd.DataFrame(data)
print(df.head())

```

```python
df['email_length'] = df['content'].apply(len)
df['word_count'] = df['content'].apply(lambda x: len(str(x).split()))
```

```python
#1D density plot

plt.figure(figsize=(7, 2))
sns.kdeplot(df['email_length'], fill=True, color='purple')
plt.title('Density Plot of Email Length')
plt.xlabel('Email Length (characters)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(4,5))
sns.kdeplot(df['word_count'],fill=True,color='blue')
plt.title('Density plot of Word count')
plt.xlabel('Word count')
plt.ylabel('Density')
plt.show()
```

```python
#2D density plot
plt.figure(figsize=(7,4))
sns.kdeplot(x=df['email_length'], y=df['word_count'], fill=True, cmap='plasma')
plt.title('2D Density Plot: Email Length vs Word Count')
plt.xlabel('Email Length')
plt.ylabel('Word Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(x=df['email_length'], y=df['word_count'], fill=False, cmap='plasma', levels=10)
plt.title('Contour Plot: Email Length vs Word Count')
plt.xlabel('Email Length')
plt.ylabel('Word Count')
plt.show()
```
